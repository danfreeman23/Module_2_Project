```python
query = '''
SELECT 
    p.user_id,
    p.device_type_canonical,
    p.order_id,
    p.created_dt_msk as order_dt,
    p.created_ts_msk as order_ts,
    p.currency_code,
    p.revenue,
    p.tickets_count,
    (p.created_dt_msk::DATE - LAG(p.created_dt_msk::DATE) OVER (
        PARTITION BY p.user_id 
        ORDER BY p.created_dt_msk::DATE
    )) AS days_since_prev,
    p.event_id,
    e.event_name_code as event_name,
    p.service_name,
    e.event_type_main,
    r.region_name,
    c.city_name
FROM afisha.purchases AS p
JOIN afisha.events AS e ON p.event_id = e.event_id
JOIN afisha.city AS c ON e.city_id = c.city_id
JOIN afisha.regions AS r ON c.region_id = r.region_id
WHERE 
    p.device_type_canonical IN ('mobile', 'desktop')
    AND e.event_type_main != '—Ñ–∏–ª—å–º'
ORDER BY p.user_id
'''

df = pd.read_sql_query(query, con=engine)
display(df)
```

```python
df_sorted = df.sort_values(['user_id', 'order_dt', 'order_ts'])
```

–î–∞–ª–µ–µ –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞ –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –º—ã —Å–æ–∑–¥–∞–µ–º –±–∞–∑–æ–≤—ã–π –ø—Ä–æ—Ñ–∏–ª—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è

```python
user_profile = df_sorted.groupby('user_id').agg({
    'order_dt': ['min', 'max'],
    'device_type_canonical': 'first',
    'region_name': 'first', 
    'service_name': 'first',
    'event_type_main': 'first',
    'order_id': 'count',
    'revenue_rub': 'mean',
    'tickets_count': 'mean',
    'days_since_prev': 'mean'
}).round(2)

user_profile.columns = [
    'first_order', 'last_order', 'first_device', 'first_region',
    'first_service', 'first_event_type', 'total_orders', 'avg_revenue',
    'avg_tickets', 'avg_days_between_orders'
]

print(f"–°–æ–∑–¥–∞–Ω–æ –ø—Ä–æ—Ñ–∏–ª–µ–π: {len(user_profile)} –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π")
```

```python
single_order_users = profiles[profiles['total_orders'] == 1]
returned_users = profiles[profiles['total_orders'] >= 2]

print(f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –æ–¥–Ω–∏–º –∑–∞–∫–∞–∑–æ–º: {len(single_order_users):,}")
print(f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2+ –∑–∞–∫–∞–∑–∞–º–∏: {len(returned_users):,}")

print("\n–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –æ–¥–Ω–∏–º –∑–∞–∫–∞–∑–æ–º:")
print(single_order_users['avg_revenue'].describe(percentiles=[0.25, 0.5, 0.75, 0.9]))
print("\n–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 2+ –∑–∞–∫–∞–∑–∞–º–∏:")
print(returned_users['avg_revenue'].describe(percentiles=[0.25, 0.5, 0.75, 0.9]))
```

–°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –≤–∏–¥–Ω—ã —Ä–∞–∑–ª–∏—á–∏—è –≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ —Å—Ä–µ–¥–Ω–µ–π –≤—ã—Ä—É—á–∫–∏ –º–µ–∂–¥—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ —Å —Ä–∞–∑–Ω—ã–º —É—Ä–æ–≤–Ω–µ–º –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏, –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø–æ–≤—Ç–æ—Ä–Ω—ã–º–∏ –∑–∞–∫–∞–∑–∞–º–∏ –∏–º–µ—é—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –±–æ–ª–µ–µ –≤—ã—Å–æ–∫—É—é —Å—Ä–µ–¥–Ω—é—é –≤—ã—Ä—É—á–∫—É, –∏ –∏—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–º–µ—â–µ–Ω–æ –≤ —Å—Ç–æ—Ä–æ–Ω—É –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏–º–∏ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –∑–∞–∫–∞–∑. –≠—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ —Ä–∞–∑–º–µ—Ä –≤—ã—Ä—É—á–∫–∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–æ–º –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–π –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, —Ç–æ–µ—Å—Ç—å –µ—Å–ª–∏ –æ–¥–∏–Ω —Ä–∞–∑ –ø–æ–ø—Ä–æ–±–æ–≤–∞–ª —Å–µ—Ä–≤–∏—Å, —Ç–æ —Ä–∞—Å—Ç–µ—Ç –Ω–∞ –∫–∞–∂–¥–æ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π

```python
users_2_4_orders = profiles[(profiles['total_orders'] >= 2) & (profiles['total_orders'] <= 4)]
users_5plus_orders = profiles[profiles['total_orders'] >= 5]

print(f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2-4 –∑–∞–∫–∞–∑–∞–º–∏: {len(users_2_4_orders):,}")
print(f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 5+ –∑–∞–∫–∞–∑–∞–º–∏: {len(users_5plus_orders):,}")

print("\n–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 2-4 –∑–∞–∫–∞–∑–∞–º–∏:")
print(users_2_4_orders['avg_revenue'].describe(percentiles=[0.25, 0.5, 0.75, 0.9]))
print("\n–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 5+ –∑–∞–∫–∞–∑–∞–º–∏:")
print(users_5plus_orders['avg_revenue'].describe(percentiles=[0.25, 0.5, 0.75, 0.9]))
```

# –ó–∞–¥–∞—á–∞ 4.2.3: –ê–Ω–∞–ª–∏–∑ –≤–ª–∏—è–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –±–∏–ª–µ—Ç–æ–≤ –Ω–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–≤—Ç–æ—Ä–Ω–æ–π –ø–æ–∫—É–ø–∫–∏


–ü–æ—Å–º–æ—Ç—Ä–∏–º –æ–±—â–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –±–∏–ª–µ—Ç–æ–≤

```python
print(profiles['avg_tickets'].describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95]))
```

–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–∏–ª–µ—Ç–æ–≤ –≤ –∑–∞–∫–∞–∑–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 2,74, –º–µ–¥–∏–∞–Ω–∞ - 2,75, 75% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ–∫—É–ø–∞—é—Ç –¥–æ 3-—Ö –±–∏–ª–µ—Ç–æ–≤ –≤ —Å—Ä–µ–¥–Ω–µ–º –∑–∞ –∑–∞–∫–∞–∑.

–ù–∞–¥–æ —Å–¥–µ–ª–∞—Ç—å —Å–µ–≥–º–µ–Ω—Ç—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –±–∏–ª–µ—Ç–æ–≤ –∏ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

```python
profiles_copy = profiles.copy()

bins = [1, 2, 3, 5, float('inf')]
labels = ['1-2 –±–∏–ª–µ—Ç–∞', '2-3 –±–∏–ª–µ—Ç–∞', '3-5 –±–∏–ª–µ—Ç–æ–≤', '5+ –±–∏–ª–µ—Ç–æ–≤']

profiles_copy['tickets_group'] = pd.cut(profiles_copy['avg_tickets'], 
                                       bins=bins, 
                                       labels=labels,
                                       right=False)

print("\n–†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–ï –ü–û–õ–¨–ó–û–í–ê–¢–ï–õ–ï–ô –ü–û –°–ï–ì–ú–ï–ù–¢–ê–ú –ë–ò–õ–ï–¢–û–í:")
segment_distribution = profiles_copy['tickets_group'].value_counts().sort_index()
total_users = len(profiles_copy)

for segment, count in segment_distribution.items():
    percentage = (count / total_users) * 100
    print(f"   {segment:<12} {count:>6,} –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ({percentage:5.1f}%)")
```

–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –Ω–µ—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ, –Ω–∞–∏–±–æ–ª—å—à–∞—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –≤ —Å–µ–≥–º–µ–Ω—Ç–µ –ø–æ 2-3 –±–∏–ª–µ—Ç–∞

–î–∞–ª–µ–µ –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –¥–æ–ª—é –≤–æ–∑–≤—Ä–∞—Ç–∞ –ø–æ –∫–∞–∂–¥–æ–º—É —Å–µ–≥–º–µ–Ω—Ç—É.

```python
grouped_analysis = profiles_copy.groupby('tickets_group').agg(
    user_count=('avg_tickets', 'count'),
    is_two_count=('is_two', 'sum'),
    is_two_mean=('is_two', lambda x: round(x.mean() * 100, 2)),
    is_five_count=('is_five', 'sum'),
    is_five_mean=('is_five', lambda x: round(x.mean() * 100, 2))
).reset_index()

print("\n–ê–ù–ê–õ–ò–ó –î–û–õ–ò –í–û–ó–í–†–ê–¢–ê –ü–û –°–ï–ì–ú–ï–ù–¢–ê–ú –ë–ò–õ–ï–¢–û–í:")
print("–°–µ–≥–º–µ–Ω—Ç       | –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π | –î–æ–ª—è 2+ –∑–∞–∫–∞–∑–æ–≤ | –î–æ–ª—è 5+ –∑–∞–∫–∞–∑–æ–≤")


for _, row in grouped_analysis.iterrows():
    print(f"{row['tickets_group']:<12} | {row['user_count']:>12,} | {row['is_two_mean']:>15}% | {row['is_five_mean']:>14}%")

avg_is_two = profiles_copy['is_two'].mean() * 100
avg_is_five = profiles_copy['is_five'].mean() * 100
print(f"\n–°—Ä–µ–¥–Ω–µ–µ –ø–æ –≤—ã–±–æ—Ä–∫–µ: | {'':>12} | {avg_is_two:>15.1f}% | {avg_is_five:>14.1f}%")
```

–°—Ç—Ä–∏–æ–º –≥—Ä–∞—Ñ–∏–∫–∏ –∏ —Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –¥–æ–ª—é –≤–æ–∑–≤—Ä–∞—Ç–∞ –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

segment_distribution.plot(kind='bar', ax=ax1, color='lightblue')
ax1.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º –±–∏–ª–µ—Ç–æ–≤')
ax1.set_xlabel('–°–µ–≥–º–µ–Ω—Ç –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –±–∏–ª–µ—Ç–æ–≤')
ax1.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
ax1.tick_params(axis='x', rotation=45)

for i, count in enumerate(segment_distribution.values):
    ax1.text(i, count, f'{count/total_users*100:.1f}%', 
             ha='center', va='bottom')

grouped_analysis.plot(x='tickets_group', y='is_two_mean', kind='bar', ax=ax2, color='lightgreen')
ax2.axhline(y=avg_is_two, color='red', linestyle='--', label=f'–°—Ä–µ–¥–Ω–µ–µ ({avg_is_two:.1f}%)')
ax2.set_title('–î–æ–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2+ –∑–∞–∫–∞–∑–∞–º–∏ –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º')
ax2.set_xlabel('–°–µ–≥–º–µ–Ω—Ç –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –±–∏–ª–µ—Ç–æ–≤')
ax2.set_ylabel('–î–æ–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2+ –∑–∞–∫–∞–∑–∞–º–∏ (%)')
ax2.tick_params(axis='x', rotation=45)
ax2.legend()

plt.tight_layout()
plt.show()
```

–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –Ω–µ—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –≤—ã—è–≤–ª—è–µ—Ç —Å–µ–≥–º–µ–Ω—Ç "2-3 –±–∏–ª–µ—Ç–∞" –∫–∞–∫ –Ω–∞–∏–±–æ–ª–µ–µ —É—Å–ø–µ—à–Ω—ã–π —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –≤–æ–∑–≤—Ä–∞—Ç–∞.

–ê–Ω–∞–ª–∏–∑ –≤—ã—è–≤–∏–ª —Å–∏–ª—å–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É —Å—Ä–µ–¥–Ω–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –±–∏–ª–µ—Ç–æ–≤ –≤ –∑–∞–∫–∞–∑–µ –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –≤–æ–∑–≤—Ä–∞—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –ù–∞–∏–±–æ–ª–µ–µ –ª–æ—è–ª—å–Ω—ã–º–∏ –æ–∫–∞–∑—ã–≤–∞—é—Ç—Å—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, –ø–æ–∫—É–ø–∞—é—â–∏–µ 2-3 –±–∏–ª–µ—Ç–∞ –∑–∞ —Ä–∞–∑ (74% –≤–æ–∑–≤—Ä–∞—Ç–∞), –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –æ—á–µ–Ω—å –º–µ–ª–∫–∏–º–∏ (1-2 –±–∏–ª–µ—Ç–∞) –∏ –æ—á–µ–Ω—å –∫—Ä—É–ø–Ω—ã–º–∏ (5+ –±–∏–ª–µ—Ç–æ–≤) –∑–∞–∫–∞–∑–∞–º–∏ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª–µ–µ –Ω–∏–∑–∫—É—é –ª–æ—è–ª—å–Ω–æ—Å—Ç—å. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º —Å–∏–ª—å–Ω–æ –Ω–µ—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Å –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–µ–π –≤ —Å–µ–≥–º–µ–Ω—Ç–µ 2-3 –±–∏–ª–µ—Ç–∞.

# –ó–∞–¥–∞—á–∞ 4.3 –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞ –∏ –∏—Ö –≤–ª–∏—è–Ω–∏—è –Ω–∞ –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –ø–æ–∫—É–ø–∫–∏

# –ó–∞–¥–∞—á–∞ 4.3.1: –ê–Ω–∞–ª–∏–∑ –≤–ª–∏—è–Ω–∏—è –¥–Ω—è –Ω–µ–¥–µ–ª–∏ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –Ω–∞ –ø–æ–≤–µ–¥–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π


–°–Ω–∞—á–∞–ª–∞ —Å–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –ø—Ä–æ—Ñ–∏–ª–µ–π –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞, –∏–∑–≤–ª–µ–∫–∞–µ–º –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –∏–∑ –¥–∞—Ç—ã –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞, –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ –¥–Ω–µ–π –Ω–µ–¥–µ–ª–∏, –≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –¥–Ω—è–º –Ω–µ–¥–µ–ª–∏

```python
profiles_copy = profiles.copy()

profiles_copy['first_order_weekday'] = profiles_copy['first_order'].dt.day_name()

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_russian = ['–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫', '–í—Ç–æ—Ä–Ω–∏–∫', '–°—Ä–µ–¥–∞', '–ß–µ—Ç–≤–µ—Ä–≥', '–ü—è—Ç–Ω–∏—Ü–∞', '–°—É–±–±–æ—Ç–∞', '–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ']

weekday_analysis = profiles_copy.groupby('first_order_weekday').agg(
    user_count=('first_order', 'count'),
    is_two_mean=('is_two', lambda x: round(x.mean() * 100, 2))
).reindex(weekday_order)

print("–†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–ï –ü–û –î–ù–Ø–ú –ù–ï–î–ï–õ–ò:")
for i, (day_eng, day_rus) in enumerate(zip(weekday_order, weekday_russian)):
    count = weekday_analysis.loc[day_eng, 'user_count']
    retention = weekday_analysis.loc[day_eng, 'is_two_mean']
    print(f"   {day_rus:<12} {count:>6,} –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ({retention:>5.1f}% –≤–æ–∑–≤—Ä–∞—Ç–∞)")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

weekday_analysis['user_count'].plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –¥–Ω—é –Ω–µ–¥–µ–ª–∏ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞')
ax1.set_xlabel('–î–µ–Ω—å –Ω–µ–¥–µ–ª–∏')
ax1.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
ax1.set_xticklabels(weekday_russian, rotation=45)

weekday_analysis['is_two_mean'].plot(kind='bar', ax=ax2, color='lightgreen')
ax2.axhline(y=profiles_copy['is_two'].mean()*100, color='red', linestyle='--', 
            label=f'–°—Ä–µ–¥–Ω–µ–µ ({profiles_copy["is_two"].mean()*100:.1f}%)')
ax2.set_title('–î–æ–ª—è –≤–µ—Ä–Ω—É–≤—à–∏—Ö—Å—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –¥–Ω—é –Ω–µ–¥–µ–ª–∏ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞')
ax2.set_xlabel('–î–µ–Ω—å –Ω–µ–¥–µ–ª–∏')
ax2.set_ylabel('–î–æ–ª—è –≤–µ—Ä–Ω—É–≤—à–∏—Ö—Å—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (%)')
ax2.set_xticklabels(weekday_russian, rotation=45)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
users_2_4_orders = profiles[(profiles['total_orders'] >= 2) & (profiles['total_orders'] <= 4)]
users_5plus_orders = profiles[profiles['total_orders'] >= 5]

print("–ê–ù–ê–õ–ò–ó –°–†–ï–î–ù–ï–ì–û –ò–ù–¢–ï–†–í–ê–õ–ê –ú–ï–ñ–î–£ –ó–ê–ö–ê–ó–ê–ú–ò:")

avg_interval_2_4 = users_2_4_orders['avg_days_between_orders'].mean()
avg_interval_5plus = users_5plus_orders['avg_days_between_orders'].mean()

print(f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 2-4 –∑–∞–∫–∞–∑–∞–º–∏: {avg_interval_2_4:.1f} –¥–Ω–µ–π")
print(f"–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 5+ –∑–∞–∫–∞–∑–∞–º–∏:  {avg_interval_5plus:.1f} –¥–Ω–µ–π")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {avg_interval_2_4 - avg_interval_5plus:.1f} –¥–Ω–µ–π ({avg_interval_2_4/avg_interval_5plus:.1f}x)")
```

# –ó–∞–¥–∞—á–∞ 4.4: –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–∫—É–ø–æ–∫ –∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è


–î–µ–ª–∞–µ–º –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞, –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏

```python
correlation_data = profiles[['total_orders', 'avg_revenue', 'avg_tickets', 'avg_days_between_orders']].copy()

interval_cols = ['avg_revenue', 'avg_tickets', 'avg_days_between_orders']

corr_matrix = correlation_data.phik_matrix(interval_cols=interval_cols)
total_orders_corr = corr_matrix['total_orders'].sort_values(ascending=False)

print("\n–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å total_orders:")
for feature, corr_value in total_orders_corr.items():
    if feature != 'total_orders':
        print(f"  {feature:<25} {corr_value:.3f}")
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.3f', 
            linewidths=0.5, center=0)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ Phi_k –¥–ª—è –≤—Å–µ—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
plt.tight_layout()
plt.show()
```

```python
users_1_order = profiles[profiles['total_orders'] == 1]
users_2_4_orders = profiles[(profiles['total_orders'] >= 2) & (profiles['total_orders'] <= 4)]
users_5plus_orders = profiles[profiles['total_orders'] >= 5]

features_for_corr = ['avg_revenue', 'avg_tickets', 'avg_days_between_orders']

segment_results = {}

print("\n–ö–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –≤ —Ä–∞–∑–Ω—ã—Ö —Å–µ–≥–º–µ–Ω—Ç–∞—Ö:")
for segment_name, segment_data in [('1 –∑–∞–∫–∞–∑', users_1_order), 
                                  ('2-4 –∑–∞–∫–∞–∑–∞', users_2_4_orders),
                                  ('5+ –∑–∞–∫–∞–∑–æ–≤', users_5plus_orders)]:
    if len(segment_data) > 0:
        segment_corr = segment_data[features_for_corr].phik_matrix(interval_cols=features_for_corr)
        segment_results[segment_name] = segment_corr
        print(f"\n{segment_name} (n={len(segment_data):,}):")
        print(segment_corr.round(3))
```

```python
categorical_features = ['first_event_type', 'first_device', 'first_region', 'first_service']
categorical_correlations = {}

for feature in categorical_features:
    temp_df = profiles[['total_orders', feature]].copy()
    corr_value = temp_df.phik_matrix()['total_orders'][0]
    categorical_correlations[feature] = corr_value

cat_corr_series = pd.Series(categorical_correlations).sort_values(ascending=False)
print("\n–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å total_orders:")
for feature, corr_value in cat_corr_series.items():
    print(f"  {feature:<20} {corr_value:.3f}")
```
